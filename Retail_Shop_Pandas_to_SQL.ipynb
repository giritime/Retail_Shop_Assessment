```python
# connection creation for oracle SQL
from sqlalchemy.engine import create_engine

DIALECT = 'oracle'
SQL_DRIVER = 'cx_oracle'
USERNAME = 'dejune22_pradhasaradhi' #enter your username
PASSWORD = 'dejune22_pradhasaradhi' #enter your password
HOST = 'orcl-aws.c8sefhobaih4.ap-south-1.rds.amazonaws.com' #enter the oracle db host url
PORT = 1521 # enter the oracle port number
SERVICE = 'ORCL' # enter the oracle db service name
ENGINE_PATH_WIN_AUTH = DIALECT + '+' + SQL_DRIVER + '://' + USERNAME + ':' + PASSWORD +'@' + HOST + ':' + str(PORT) + '/?service_name=' + SERVICE

engine = create_engine(ENGINE_PATH_WIN_AUTH)
```

```python
# Create a table in Oracle database Via Pandas
try:
    connection = cx_Oracle.connect('dejune22_pradhasaradhi/dejune22_pradhasaradhi@orcl-aws.c8sefhobaih4.ap-south-1.rds.amazonaws.com:1521/ORCL')
    print(connection.version)
    
    # Now execute the sqlquery
    cursor = connection.cursor()
    
    #importing CSV file to Pandas dataframe
    Country_df = pd.read_csv(r'D:\Learning\iBridge 360\Data Warehouse Concepts\My Project\CSV_files\Country.csv')
    print("Table Successfully Imported to Pandas")
    
    
    #Uploading Pandas dataframe to SQL   
    Country_df.to_sql('country_r', con = engine, if_exists = 'append', index=False)
    #Read From SQL Table
    pd.read_sql("SELECT * FROM country_r", con = engine)
    print("Appending Entries Successfully")
    
    cursor.execute("COMMIT")
    
except cx_Oracle.DatabaseError as e:
    print("There is a problem with Oracle", e)
 
# by writing finally if any error occurs
# then also we can close the all database operation
finally:
    if cursor:
        cursor.close()
    if connection:
        connection.close()
```
